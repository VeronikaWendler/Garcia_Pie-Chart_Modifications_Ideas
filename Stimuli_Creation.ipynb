# Stimuli description

##### Stimuli are presented as pie-charts. We are interested in assessing whether participants show the same experiential neglect as previously seen in Garcia et al's (2022/23) study on neural representations of value representation and reward processing. The heading indicates the green/red ratio. Each code additionally generates the inverse ratio. There are the factors: Folding condition and distance. The levles of the distance condition are 'equal distance' and 'random distance'. For 'equal distance', the subsegments are distributed in equal distances within the pie chart. For 'random distance', the subsegments are randomly distributed within the pie chart. There are 6 levels of folding. Green or red is split into 3, 5, 7, 9, or 11 subsegments and interleaved in either equal distance or randomly. Pie-charts are generated randomly; for each run a unique angle (drawn from unifrom distribution) and/or distance between the subsegments is generated.
